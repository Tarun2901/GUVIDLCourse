```python
import numpy as np
import matplotlib.pyplot as plt
```

# Vector Plotting


```python
# Quiver function is used to plot vectors
plt.quiver(0,0,3,4)
plt.show()
```

```python
# Quiver function used to plot vectors (Corrected scale)
plt.quiver(0,0,3,4,scale_units = 'xy', angles = 'xy', scale = 1, color = 'red')
plt.quiver(0,0,-3,4,scale_units = 'xy', angles = 'xy', scale = 1, color = 'green')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

```python
# Write a function to plot all vectors from a given list of vectors in one go

def plot_vectors(vecs):
  colors = ['r','b','g','y']
  i = 0
  for vec in vecs:
    plt.quiver(vec[0],vec[1],vec[2],vec[3],scale_units = 'xy', angles = 'xy', scale = 1, color = colors[i%len(colors)])
    i = i + 1
  plt.xlim(-10,10)
  plt.ylim(-10,10)
  plt.show()
```

```python
plot_vectors([(0,0,3,4),(0,0,-3,4),(0,0,-3,-2),(0,0,4,-1)])
```

```python
vecs = [np.asarray([0,0,3,4]),np.asarray([0,0,-3,4]),np.asarray([0,0,-3,-2]),np.asarray([0,0,4,-1])]
```

```python
plot_vectors([vecs[0],vecs[3]])
```

# Vector dot product

```python
vecs = [np.asarray([0,0,3,4]),np.asarray([0,0,-3,4]),np.asarray([0,0,-3,-2]),np.asarray([0,0,4,-1])]
```

```python
plot_vectors(vecs)
```

```python
a = np.asarray([3,4])
b = np.asarray([4,-1])
```

$\vec{a}\cdot\vec{b} = |\vec{a}| |\vec{b}| \cos(\theta) = a_x b_x + a_y b_y $

```python
a_dot_b = np.dot(a,b)
```

```python
print(a_dot_b)
```

$a_b = |\vec{a}| \cos(\theta) = |\vec{a}| \frac{\vec{a} \cdot \vec{b}}{|\vec{a}| |\vec{b}|} = \frac{\vec{a} \cdot \vec{b}}{|\vec{b}|} $

```python
a_b = np.dot(a,b)/np.linalg.norm(b)
```

```python
print(a_b)
```

$\vec{a_b} = a_b \hat{b} = a_b \frac{\vec{b}}{|\vec{b}|} $

```python
# Projections's direction will be along b
vec_a_b = (a_b/np.linalg.norm(b))*b
print(vec_a_b)
```
